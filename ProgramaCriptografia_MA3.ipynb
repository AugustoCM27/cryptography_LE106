```
# Criptografia

# MA3 - 24/06/2020
# LE106 - GAAL - Prof.: Priscila Cristina Berbert Rampazzo

# Integrantes do Grupo:
# Augusto Chebel Machado          RA: 231762
# Nathan Czelusniak de Siqueira   RA: 186041 
# Nicolas Keichi Matsuse Shirai   RA: 186143 
# Vitor Janerilo Cardoso          RA: 195228

# Codificação (1)

#
print("\033[0;34m#"*45)
print("\033[0;31mPROGRAMA DE CODIFICAÇÃO DA MELHOR DISCIPLINA")
print("\033[0;34m#"*45)
print()
#

# Etapa 01:
frase = str(input("Digite a frase que você deseja criptografar: ")).strip()         # Digitar a frase 
listafrase = []                                                                     
for letra in frase:                                                                 # Relacionar cada letra ou espaço a um número 
  if letra == " ":
    numero = 0
  else:
    numero = ord(letra) - 96
  listafrase.append(numero)

# Etapa 02:
termos = len(listafrase)      # Fazer com que o número de termos seja divisível por 3
if termos % 3 == 0:           # Objetivo: matrizes com 1 linha e 3 colunas
  listafrase = listafrase
elif termos % 3 == 1:
  listafrase.append(0)
  listafrase.append(0)
elif termos % 3 == 2:
  listafrase.append(0)

# Etapa 03:
import numpy as np

matrizmult = np.array([listafrase[i : i+3] for i in range(0 , len(listafrase) , 3)])    # Divide a lista em linhas de 3 termos e coloca-se em uma matriz

# Etapa 04:
A = np.array([[1 , -2 , 2] ,                # Define-se a matriz que será a chave da criptografia
              [-1 , 1 , 3] ,
              [1 , -1 , -4]])

# Etapa 05:
matrizfinal = np.matmul(matrizmult , A)     # Faz-se a multiplicação entre as linhas e matriz-chave

# Etapa 06:
listacodigo = []                                  # Código em forma de lista
for i in range(len(matrizfinal)):
  for j in range(3):
    listacodigo.append(matrizfinal[i][j])

#
from time import sleep
print()
print("Fazendo a codificação...")
sleep(2)
print("Aguarde um instante...")
sleep(2)
print("Finalizando...")
sleep(1)
#

# Etapa 07:
print()
print("A mensagem codificada é: ")              # Mostra a mensagem codificada 
for i in range(len(listacodigo)):               # No formato de código, sem vírgulas ou chaves
  print(listacodigo[i], end=' ')
print("\n")
print("Obrigado por utilizar nosso programa!")
```

```
# Decodificação (2)

#
print("\033[0;34m#"*48)
print("\033[0;31mPROGRAMA DE DECODIFICAÇÃO DA MELHOR DISCIPLINA")
print("\033[0;34m#"*48)
print()
#

# Etapa 01:
import numpy as np

codigo = input("Insira o código para revelar o conteúdo da mensagem: ").split()    # Colocar o código para descriptografar

# Etapa 02:
for i in range(len(codigo)):     # Transforma os elementos em inteiros
  codigo[i] = int(codigo[i])

# Etapa 03:
matrizreversa = np.array([codigo[i : i+3] for i in range(0 , len(codigo) , 3)])    # Transforma-se os elementos em uma matriz de linhas com 3 termos

# Etapa 04:
A = np.array([[1 , -2 , 2] ,        # Define-se A novamente para que não seja necessário rodar o programa de cripotografia     
              [-1 , 1 , 3] ,        # Programas independentes
              [1 , -1 , -4]])

invA = np.linalg.inv(A)             # Define-se a matriz inversa (em relação à matriz-chave A)

# Etapa 05:
matrizoriginal = np.matmul(matrizreversa , invA)     # Multiplicação

# Etapa 06:
lista = []                                         # Cria-se uma lista com os valores obtidos através da multiplicação
for i in range(len(matrizoriginal)):               # Passa-se por cada linha e coluna
  for j in range(3):
    lista.append(matrizoriginal[i][j])

# Etapa 07: 
listadecod = []                                  # Transforma cada elemento da lista em inteiros
for c in range(0 , len(lista) , 1):
  elemento = int(lista[c])
  listadecod.append(elemento)

# Etapa 08:
mensagemoriginal = []             # Conversão dos números em letras , de acordo com a tabela ascii
for numero in listadecod:         
  if numero == 0:
    letra = " "
  else:
    letra = chr(numero + 96)
  mensagemoriginal.append(letra)

#
from time import sleep
print()
print("Analisando padrões...")
sleep(2)
print("CARREGANDO...")
sleep(2)
print()
#

# Etapa 09:
mensagemfinal = "".join(mensagemoriginal)                   # Junta as letras para formar a frase original
print("A mensagem original era:")                           
print(mensagemfinal)
print()
print("Obrigado por utilizar nosso programa!")
```
